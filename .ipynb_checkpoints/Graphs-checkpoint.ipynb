```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 20)

plt.plot(x, (3*(x**2)) + (20*x), label='i')
#plt.plot(x, (x**3) - (3*(x**2)) + (5*x) + 8, label='ii')

plt.xlabel('x label')
plt.ylabel('y label')

plt.scatter((-10/3),(-100/3),color='#ff0000')

plt.title("i")

plt.legend()

plt.show()
```
